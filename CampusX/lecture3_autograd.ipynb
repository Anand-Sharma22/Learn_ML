```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-7  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)  # Clamp predictions to avoid log(0)
    loss = - (target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
    return loss
```

```python
# forward pass
z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
loss
```
