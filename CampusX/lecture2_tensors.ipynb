### 3. reduction operations

```python
e = torch.randint(size=(2,3), low=0, high=10)
e
```

```python
# sum
e.sum()
torch.sum(e)
```

```python
# sum along the columns
torch.sum(e, dim=0)
# sum along the rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e.float()) 
# mean along the columns
torch.mean(e.float(), dim=0)
# mean along the rows
torch.mean(e.float(), dim=1)
```

```python
# median 
torch.median(e.float())
# max and min
torch.max(e)
torch.min(e)
```

```python
# product
torch.prod(e)
```

```python
# standard deviation
torch.std(e.float())
```

```python
# variance
torch.var(e.float())
```

```python
# argmax and argmin
torch.argmax(e)
torch.argmin(e)
```

### 4. Matrix Operations

```python
f = torch.randint(size=(2,3), low=0, high=10, dtype = torch.float32)
g = torch.randint(size=(3,2), low=0, high=10, dtype = torch.float32)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
# dot product
torch.dot(f.flatten(), g.flatten())
```

```python
# transpose
torch.transpose(f, 0, 1)
```

```python
# determinant
torch.det(torch.rand(3,3))
# inverse
torch.inverse(torch.rand(3,3))
```

### 5. Comparision operation

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)
print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than or equal to
i >= j
# less than or equal to
i <= j
```

### 6. Special Functions

```python
k = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
```

## Tensor Operations on GPU (MPS)

```python
torch.mps.is_available()
```

```python
device = torch.device("mps") if torch.mps.is_available() else torch.device("cpu")
device
```

```python
# Creating new tensor on GPU/MPS
torch.rand(2,3, device=device)
```

```python
# moving an existing tensor to GPU/MPS
x = torch.rand(2,3)
print(x)
b = x.to(device)
```

```python
b + 5
```

```python
import time
size = 10000

# CPU
Matrix1_CPU = torch.randn(size, size)
Matrix2_CPU = torch.randn(size, size)

# Time on CPU 
start_time = time.time()
result_CPU = torch.matmul(Matrix1_CPU, Matrix2_CPU)
end_time = time.time()
print(f"Time taken on CPU: {end_time - start_time:.4f} seconds")

# GPU/MPS
matrix1_MPS = Matrix1_CPU.to(device)
matrix2_MPS = Matrix2_CPU.to(device)

# Time on GPU/MPS
start_time = time.time()
result_MPS = torch.matmul(matrix1_MPS, matrix2_MPS)
end_time = time.time()
print(f"Time taken on GPU/MPS: {end_time - start_time:.4f} seconds")
```
